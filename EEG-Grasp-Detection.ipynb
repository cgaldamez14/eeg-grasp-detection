```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

data = []
events = []
for series in range(1,9):
    d = pd.read_csv("train/subj1_series" + str(series)+ "_data.csv",index_col=None, header=0)
    e = pd.read_csv("train/subj1_series" + str(series)+ "_events.csv",index_col=None, header=0)
    data.append(d)
    events.append(e)

dataConcat = pd.concat(data,ignore_index=True)
eventConcat = pd.concat(events,ignore_index=True)

X = dataConcat[["Fp1", "Fp2", "F7", "F3", "Fz", "F4", "F8", "FC5", "FC1", "FC2","FC6"]]
#X = dataConcat.drop("id", axis = 1)
y = eventConcat.drop("id", axis = 1)

%matplotlib inline

plt.figure();
plt.plot(X["Fp1"])
plt.show()

plt.figure();
plt.plot(X["F7"])
plt.show()

plt.figure();
plt.plot(X["FC1"])
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neural_network import MLPClassifier

my_ANN = MLPClassifier(hidden_layer_sizes=(10,4), activation= 'logistic', 
                       solver='adam', alpha=1e-5, random_state=1, 
                       learning_rate_init = 0.1)

accuracy_list = cross_val_score(my_ANN, X, y, cv=5, scoring='accuracy')
```

```python
print(accuracy_list)
print(accuracy_list.mean())
```
